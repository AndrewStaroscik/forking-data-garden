```python
import random
import numpy as np
import matplotlib.pyplot as plt

```

```python

random.seed(92)

def sampWithReplacement(opts, trials=1000, draws=5):
  yesCount = 0

  for el in range(trials): 
    success = 0
    for draw in range(draws):
      d = random.choice(opts)
      if d == 'O':
        success = 1
    if success == 1:
      yesCount += 1

  return round(100 * yesCount / trials, 1)

options = ['O', 'G', 'G', 'G', 'G']
results = []

for t in range(0,21,1):
  results.append(sampWithReplacement(options, int(1e6), t))
                 
results
```

```python
xLabels = [5, 10, 15, 20]

fig, ax = plt.subplots()



ax.plot(np.arange(len(results)), results, color='orange')
ax.set_xticks(xLabels)
ax.set_xlabel('Draws')
ax.set_ylabel('Probability of at least one orange draw (%)')
ax.set_xlabel('Draws')
ax.set_ylabel('Probability of at least one Orange (%)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
